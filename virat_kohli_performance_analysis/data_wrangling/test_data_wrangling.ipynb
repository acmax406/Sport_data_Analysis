```python
import numpy as np
import pandas as pd
import glob
import os
```

```python
# get data file names
path =r'C:\Users\Anand\Documents\sports\raw_dataset\test_dataset'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    try:
        data = pd.read_csv(filename)
    #     print(data)
        dfs.append(data)
    except:
        dfs.append(pd.read_csv(r'C:/Users/Anand/Documents/sports/empty.csv'))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
print(big_frame)
```

```python
df = big_frame[big_frame['0'].str.contains("Virat Kohli")]
df = df.reset_index()
df
```

```python

df['out_type'] = df.out_by.apply(lambda x: x[:1])
df['out_type'] = df['out_type'].str.replace('l','lbw')
df['out_type'] = df['out_type'].str.replace('c','catch_out')
df['out_type'] = df['out_type'].str.replace('r','run_out')
df
```

```python
df['out_by'] = df['out_by'].str.replace('lbw b','')
df['out_by'] = df['out_by'].str.replace('run out','')
df['out_by'] = df['out_by'].str.replace('b','')
df['out_by'] = df['out_by'].str.replace('c †','')
df['out_by'] = df['out_by'].str.replace('c','')

df
```

```python
data = df['out_by'].str.rsplit(' ',n=1,expand= True)
df['outby']=data[1]
df= df.drop(['out_by'],axis =1)
df
```

```python
df.to_csv(r'test_record.csv',index=False)
```
