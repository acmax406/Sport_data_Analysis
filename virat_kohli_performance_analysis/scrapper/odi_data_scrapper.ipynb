```python

from bs4 import BeautifulSoup
import requests
from urllib.parse import urlparse, parse_qs
import re
import pandas as pd
import numpy as np
 
webpage = requests.get('https://stats.espncricinfo.com/ci/engine/player/253802.html?class=2;template=results;type=allround;view=match')
sp = BeautifulSoup(webpage.content, 'html.parser')
# print(sp.text)
```

```python
titles = sp.find_all(title='view the scorecard for this row')

def dataset(title):
    print(title)

    webpage2 = requests.get(title)
    sp2= BeautifulSoup(webpage2.content,'html.parser')
    
    table = sp2.findAll("table", { "class" : "ds-w-full ds-table ds-table-xs ds-table-fixed ci-scorecard-table" })
    
#     print(table[0].text)
#     print(table[1].text)
    
    player_name=[]
    out_by =[]
    runs=[]
    balls=[]
    empty=[]
    fours =[]
    sixes=[]
    strike=[]
    
    if len(table)>1:
        i = 0
        for row in table[0].findAll("tr"):
            if i%2 ==0:
                cells = row.findAll("td")
                temp = len(cells)
                if temp%8==0:
                    j=0
                    for cell in cells:
    #                     print(cell.text)
                        if j==0:
                            player_name.append(cell.text)
                            j=j+1
                        elif j==1:
                            out_by.append(cell.text)
                            j=j+1
                        elif j ==2:
                            runs.append(cell.text)
                            j=j+1
                        elif j ==3:
                            balls.append(cell.text)
                            j=j+1
                        elif j ==4:
                            empty.append(cell.text)
                            j=j+1
                        elif j ==5:
                            fours.append(cell.text)
                            j=j+1
                        elif j ==6:
                            sixes.append(cell.text)
                            j=j+1
                        else:
                            strike.append(cell.text)
                            j=j+1

            else:
                i =i+1

        k = 0
        for row in table[1].findAll("tr"):
            if k%2 ==0:
                cells = row.findAll("td")
                temp = len(cells)
                if temp%8==0:
                    m=0
                    for cell in cells:
    #                     print(cell.text)
                        if m==0:
                            player_name.append(cell.text)
                            m=m+1
                        elif m==1:
                            out_by.append(cell.text)
                            m=m+1
                        elif m ==2:
                            runs.append(cell.text)
                            m=m+1
                        elif m ==3:
                            balls.append(cell.text)
                            m=m+1
                        elif m ==4:
                            empty.append(cell.text)
                            m=m+1
                        elif m ==5:
                            fours.append(cell.text)
                            m=m+1
                        elif m ==6:
                            sixes.append(cell.text)
                            m=m+1
                        else:
                            strike.append(cell.text)
                            m=m+1

            else:
                k =k+1


        df = pd.DataFrame(player_name)
        df['out_by']=out_by
        df['runs']=runs
        df['balls']=balls
        df['empty']=empty
        df['fours']=fours
        df['sixes']=sixes
        df['strike_rate']=strike


        return df
    else:
        df2 = pd.DataFrame()
        return df2
```

```python
for title in titles:
    print(title.text)
    name= title.text
    title = title.get('href')
    
#    /ci/engine/match/343732.html 

    title = 'https://stats.espncricinfo.com'+title
    test_data= dataset(title)
    
    test_data.to_csv(r'odi_n'+name+'.csv',index =False)
    
```
